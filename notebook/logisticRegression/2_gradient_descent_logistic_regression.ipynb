# Gradient Descent for Logistic Regression

```python
import copy, math
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
def plot_data(X, y, ax):
    positive = y == 1
    negative = y == 0
    ax.scatter(X[positive, 0], X[positive, 1], marker = 'x', color ='red', label = 'y = 1' )
    ax.scatter(X[negative, 0], X[negative, 1], marker = 'o', color = 'blue', label = 'y = 0')
    


fig, ax = plt.subplots(1, 1, figsize = (4,4))
plot_data(X_train, y_train, ax)
ax.axis([0, 4.0, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize = 12)
ax.set_xlabel('$x_0$', fontsize = 12)
plt.show()


```

```python
def sigmoid(z) :
    ans = 1 / (1 + np.exp(-z))
    return ans;
    
```

```python
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]
    z = np.dot(X, w) + b
    f_wb = sigmoid(z)
    error = f_wb - y
    cost = (np.sum(error**2)) / (2*m)
    return cost
```

## Gradient Descent Implementation

```python
def compute_gradient_logistic_scaler(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.0

    print(dj_dw)

    
    for i in range(m):
        z = np.dot(X[i] , w ) + b
        f_wb_i = sigmoid(z)
        error = f_wb_i - y[i]

        dj_dw += np.dot(error, X[i].T)
        dj_db += error

    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db  
        
```

```python
def compute_gradient_logistic_vector(X, y, w, b):
    m = X.shape[0]
    z = np.dot(X, w) + b
    f_wb = sigmoid(z)
    error = f_wb - y

    dj_dw = np.dot(X.T, error)/m
    dj_db = np.sum(error)/m

    return dj_dw, dj_db

    
```

### Gradient Descent code


```python
def gradient_descent(X, y, w_init, b_init, alpha, iteration):

    J_history =[] #cost history
    w = copy.deepcopy(w_init) # avoid modifying global w
    b = b_init

    for i in range (iteration) :

        dj_dw, dj_db = compute_gradient_logistic_vector(X, y, w, b)

        w = w - alpha*dj_dw
        b = b - alpha*dj_db
        
        if( i < 10000):
            J_history.append(compute_cost_logistic(X, y, w, b))

        if i% math.ceil(iteration / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")


    return w, b, J_history


    
```

```python
w_init = np.array([0,0])
b_init = 0;
alpha = 0.1
iter = 10000
w, b, J_history = gradient_descent(X_train, y_train, w_init, b_init, alpha, iter)
print(f"w : {w}, b: {b}")
```
